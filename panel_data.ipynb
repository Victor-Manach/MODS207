```python
import numpy as np
import pandas as pd
```

```python
scraping_date1, scraping_date2 = '_26_04_2021', '_17_05_2021'
df1 = pd.read_csv('./data/full_profiles'+scraping_date1+'.csv')
df2 = pd.read_csv('./data/full_profiles'+scraping_date2+'.csv')
df1['scraping_date'] = len(df1)*['26-04-2021']
df2['scraping_date'] = len(df2)*['17-05-2021']
```

```python
names1, names2 = np.array(df1.name), np.array(df2.name)
genders = []
for name in names2:
    if name in names1:
        gender = df1[df1.name==name]['gender'].values[0]
        genders.append(gender)
    else:
        genders.append('unknown')
df2 = df2.assign(gender=genders)
```

```python
panel_data = pd.DataFrame(columns=df1.columns)
lost_profiles = pd.DataFrame(columns=df1.columns)
new_profiles = pd.DataFrame(columns=df1.columns)
```

```python
for name in names2:
    if name in names1:
        panel_data = panel_data.append(df1[df1.name==name], ignore_index=True)
        panel_data = panel_data.append(df2[df2.name==name], ignore_index=True)
    else:
        new_profiles = new_profiles.append(df2[df2.name==name])

for name in names1:
    if name not in names2:
        lost_profiles = lost_profiles.append(df1[df1.name==name])
```

```python
new_profiles, lost_profiles = new_profiles.reset_index(drop=True), lost_profiles.reset_index(drop=True)
```
