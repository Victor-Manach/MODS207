```python
import numpy as np
import pandas as pd
import time
import utils
```

```python
today = time.strftime("_%d_%m_%Y")
data = pd.read_csv('./data/profiles_info'+today+'.csv')
```

# Determine the gender using the name of the person

```python
nan_gender_profiles = data[data.gender.isna()]
```

```python
nan_gender_profiles
```

```python
genders = []
for i, profile in enumerate(nan_gender_profiles.itertuples()):
    full_name = profile.name
    fname = full_name.split()[0]
    gender = utils.guess_gender_from_name(fname)
    genders.append(gender)
    if i%500==0:
        print('%d profiles checked' % i)
print('Finished...')
```

```python
nan_gender_profiles = nan_gender_profiles.assign(gender=genders)
```

```python
nan_gender_profiles
```

```python
'Determined the gender of {}% of the {} profiles'.format((1-len(nan_gender_profiles[nan_gender_profiles.gender.isna()])/len(nan_gender_profiles))*100, len(nan_gender_profiles))
```

```python
data[data['gender'].isna()] = nan_gender_profiles
```

```python
gender = data.gender
perc = (len(gender[gender=='male']) + len(gender[gender=='female']))/len(gender)*100
'Percentage of profiles where we have determined the gender {}%'.format(round(perc, 2))
```

```python
data.to_csv('./data/improved_profiles'+today+'.csv', index=False)
```
