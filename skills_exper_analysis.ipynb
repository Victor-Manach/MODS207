```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re, ast
from collections import Counter
import seaborn as sn

sn.set()
```

```python
scraping_date = '_26_04_2021'
df = pd.read_csv('./data/full_profiles'+scraping_date+'.csv')
```

```python
skills = df.skills
list_of_skills = []
for profile_skills in skills:
    profile_skills = ast.literal_eval(profile_skills)
    for skill in profile_skills:
        list_of_skills.append(skill)
```

```python
Counter(list_of_skills).most_common(20)
```

```python
def find_translation_skills(profile_skills: list) -> list:
    for skill in profile_skills:
        translation = re.search(r'\btranslation\b',skill)
        print(translation)
```

# Find most active profiles

```python
active = df[df.nb_reviews>100].reset_index(drop=True)
```

```python
figure = plt.figure(figsize=(15,8))

axes = figure.add_subplot(121)
sn.countplot(x='gender', data=active)
plt.title("Gender distribution among the most active profiles", fontsize = 18)


axes = figure.add_subplot(122)
axes.pie(active["gender"].value_counts(),radius = 10, autopct='%1.1f%%', textprops=dict(color="black"))
axes.axis('equal')
```

```python
list_of_skills = []
for profile_skills in active.skills:
    profile_skills = ast.literal_eval(profile_skills)
    for skill in profile_skills:
        list_of_skills.append(skill)
```

```python
list_of_expertise = []
for profile_expertise in active.industry_expertise:
    profile_expertise = ast.literal_eval(profile_expertise)
    for expertise in profile_expertise:
        list_of_expertise.append(expertise)
```

```python
Counter(list_of_skills).most_common(20)
```
